```python
import pandas as pd
sp=pd.read_csv('/Users/xuzhichao/Desktop/TSLA_3years.csv',index_col=0, parse_dates = True)
#sp.set_index('Date',inplace=True)
sp.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,5))
sp['Close'].plot(title='TSLA',linewidth=0.7)
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()
```

```python
sp['2020']
```

```python
plt.scatter(sp.Volume,sp.Close, marker = '^', alpha = 0.8, c = sp.index)
plt.grid(False)
plt.colorbar()
plt.title('Close Price VS Volume')
plt.ylabel('Close Price ($)')
plt.xlabel('Volume')
plt.show()
```

```python
sp_2020= sp['2020'].resample('M').mean()
sp_2020
```

```python
plt.bar(sp_2020.index.month, sp_2020.Volume, color = 'SkyBlue', hatch='o')
plt.xlabel('Month')
plt.ylabel('Volume')
plt.title('Active Trading Month')
plt.show()
```

```python
sp_year= sp.resample('Y').mean()
sp_year
```

```python
plt.bar(sp_year.index.year, sp_year.Volume, color = 'SkyBlue', hatch='o')
plt.xlabel('Year')
plt.ylabel('Volume')
plt.title('Active Trading Year')
plt.show()
```

```python
sp_2020
```

```python
fig, ax = plt.subplots()
label = ['Open', 'Close', 'High', 'Low']
ax.boxplot(x = sp_2020[label].values, labels = label)
ax.set_title('Price Distribution in 2020', fontsize = 16)
plt.grid()
plt.show()
```
